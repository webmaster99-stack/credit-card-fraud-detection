## Project Conclusion

### General Comments

This project attempted to solve the classification problem of detecting credit card fraud using machine learning techniques. The project also compares models trained on synthenic transaction data and real-word transaction data from european cardholders.

Multiple algirithms were tested on both dataset but the best performing on both datasets was the Gradient Boosting Classifer.
